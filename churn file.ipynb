```python
x=df.drop(['customer_id','churn',],axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
lb_encod=LabelEncoder()
```

```python
x["gender"]=lb_encod.fit_transform(df["gender"])
```

```python
x
```

```python
ohe=OneHotEncoder(sparse_output=False)
```

```python
geo_encoded=ohe.fit_transform(df[["country"]])
```

```python
ohe.get_feature_names_out()
```

```python
new_df=pd.DataFrame(geo_encoded,columns=ohe.get_feature_names_out())
```

```python
new_df
```

```python
x=pd.concat([x,new_df],axis=1)
```

```python
x
```

```python
x.drop(["country",'Gender'],axis=1,inplace=True)
```

```python
x["age"]=df["age"].astype(int)
```

```python
y=df["churn"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
param_grid={
    'C':[0.01,0.1,1,10],
    'penalty':['l1','l2'],
    'solver':['liblinear']
}
```

```python
grid=GridSearchCV(
    estimator=LogisticRegression(),
    param_grid=param_grid,
    scoring="accuracy"
)
```

```python
grid.param_grid
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.param_grid
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
print("Logistic \n",classification_report(y_test,grid.predict(x_test)))
print("Logistic \n",classification_report(y_train,grid.predict(x_train)))
```

```python
param_grid={
    'max_depth':[6,10,15],
    'min_samples_split':[2,5],
    'min_samples_leaf':[1,2]
}
```

```python
gridd=GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=param_grid,
    scoring="accuracy"
)
```

```python
gridd.fit(x_train,y_train)
```

```python
gridd.best_params_
```

```python
gridd.best_estimator_
```

```python
gridd.best_score_
```

```python
print("Unseen DecisionTreeClassifier \n",classification_report(y_test,gridd.predict(x_test)))
print("Seen DecisionTreeClassifier \n",classification_report(y_train,grid.predict(x_train)))
```

```python
param_grid_rf={
    'n_estimators':[50,100],
    'max_depth':[5,10,20],
    'min_samples_split':[2,5],
    "min_samples_leaf":[1,2]
}
```

```python
grid_rf=GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=param_grid_rf,
    scoring="accuracy"
)
```

```python
grid_rf.fit(x_train,y_train)
```

```python
grid_rf.best_estimator_
```

```python
grid_rf.best_params_
```

```python
grid_rf.best_score_
```

```python
print("Unseen rf  \n",classification_report(y_test,grid_rf.predict(x_test)))
print("Seen rf \n",classification_report(y_train,grid.predict(x_train)))
```

```python
param_grid_gb={
    'n_estimators':[50,100],
    "max_depth":[5,10],
    "learning_rate":[0.05,0.1],
    'min_samples_split':[2,5],
    'min_samples_leaf':[1,2],
    
}
```

```python
grid_gb=GridSearchCV(
    estimator=GradientBoostingClassifier(),
    param_grid=param_grid_gb,
    scoring='accuracy'
)
```

```python
grid_gb.fit(x_train,y_train)
```

```python
grid_gb.best_estimator_
```

```python
grid_gb.best_score_
```

```python
grid_gb.best_params_
```

```python
print("Unseen GradientBoosting \n",classification_report(y_test,grid_gb.predict(x_test)))
print("Seen GradientBoosting \n",classification_report(y_train,grid_gb.predict(x_train)))
```
